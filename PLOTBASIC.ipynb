```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_seconds = pd.read_csv('WSlog_202011.csv')
df_minutes = pd.read_csv('WSlog_202011.csv')
df_seconds.head()
```

```python
df_seconds.columns = ['Time', '0.0', '0.0.1', '0.0.2', '0.0.3', 'Temperature1', 'Temperature2']
df_minutes.columns = ['Time', '0.0', '0.0.1', '0.0.2', '0.0.3', 'Temperature1', 'Temperature2']
```

```python
def convert_to_seconds(time):
    hours, minutes, seconds = time.split(':')
    converted_time = (int(hours) * 60 * 60) + (int(minutes) * 60) + int(seconds)
    return converted_time

df_seconds['Time'] = df_seconds['Time'].apply(convert_to_seconds)
```

```python
def convert_to_minutes(time):
    hours, minutes, seconds = time.split(':')
    converted_time = (float(hours) * 60) + float(minutes) + (float(seconds) / 60)
    return converted_time

df_minutes['Time'] = df_minutes['Time'].apply(convert_to_minutes)
```

```python
df_seconds.plot(kind='line', x='Time', y='Temperature1', color='red')
plt.show()
```

```python
df_minutes.plot(kind='line', x='Time', y='Temperature1', color='red')
plt.show()
```
