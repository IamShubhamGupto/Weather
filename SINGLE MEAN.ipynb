```python
import pandas as pd
from statistics import mean, median, mode, stdev
f = open('WSlog_2020323.csv','r')
data = f.readlines()
answerlist = []
i=0
while i < 480:
        first = float(data[i].split(',')[1])
        second = float(data[i+1].split(',')[1])
        temp = (first+second)/2
        answerlist.append(temp)
        i = i + 2
        print(answerlist, len(answerlist))
        f.close()
```
