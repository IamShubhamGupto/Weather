```python
from statistics import mean, median, mode, stdev
f = open('WSlog_202056.csv','r')
data = f.readlines()
#data = data[1:]
final_list = []

length = len(data)
j = 0
while j < 480:
    i = j
    six_list = []
    print(i)
    for _mode in range(1):    
        first = float(data[i].split(',')[2])
        second = float(data[i+1].split(',')[2])
        temp = (first+second)/2
        six_list.append(temp)
        i += 480
    mean_162 = mean(six_list)
    final_list.append(mean_162)
    j += 2
print(final_list, len(final_list))

f.close()
```
